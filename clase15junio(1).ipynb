```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
df1_x = np.asanyarray(df1[['Monto']])
df1_y = np.asanyarray(df1[['Tasa']])
regr.fit (df1_x, df1_y)
# The coefficients
print ('coefficients:',regr.coef_)
print ('intercept: ',regr.intercept_)
```

```python
plt.scatter(df1.Monto, df1.Tasa, color='blue')
plt.plot(df1_x, regr.coef_[0][0]*df1_x + regr.intercept_[0], '-r')
plt.xlabel("Monto")
plt.ylabel("Tasa")
```
